```python
import pandas as pd
path ="C:/Users/Abdoulie Balisa/Downloads/airline_dataset.csv"
airlines = pd.read_csv(path)
airlines.head(5)
```

```python
categories = {
    "day": {"Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"},
    "airline": {"Airline A", "Airline B", "Airline C", "Airline D"},
    "destination": {"New York", "London", "Paris", "Tokyo", "Dubai"},
    "dest_region": {"North America", "Europe", "Asia", "Middle East"},
    "dest_size": {"Small", "Medium", "Large"},
    "boarding_area": {"Gate A", "Gate B", "Gate C", "Gate D"},
    "dept_time": set(range(0, 24)),  # Hours (0 to 23)
    "wait_min": set(range(5, 181)),  # Waiting time (5 to 180 minutes)
    "cleanliness": {"Good", "Excellent", "Poor"},  # Valid cleanliness values
    "safety": {"Safe", "Moderate", "Unsafe"},
    "satisfaction": {"Satisfied", "Neutral", "Dissatisfied"},
}
```

```python
cat_clean = set(airlines['cleanliness']).difference(categories['cleanliness'])
cat_clean
```

```python
# Find rows with that category
cat_clean_rows = airlines[airlines['cleanliness'].isin(cat_clean)]
cat_clean_rows.head()
```

```python
valid_clean_rows = airlines[~airlines['cleanliness'].isin(cat_clean)]
valid_clean_rows.head()
```

```python
# Replace "Dr." with empty string ""
airlines['full_name'] = airlines['full_name'].str.replace("Dr.","")

# Replace "Mr." with empty string ""
airlines['full_name'] = airlines['full_name'] = airlines['full_name'].str.replace("Mr.","")


# Replace "Miss" with empty string ""
airlines['full_name'] = airlines['full_name'].str.replace("Miss","")


# Replace "Ms." with empty string "
airlines['full_name'] =airlines['full_name'].str.replace("Ms.","")


# Assert that full_name has no honorifics
assert airlines['full_name'].str.contains('Ms.|Mr.|Miss|Dr.').any() == False
```
