```python
from bs4 import BeautifulSoup
import pandas as pd
import requests

# Base URL
url = 'https://www.who.int/health-topics/#A'

# Send GET request
response = requests.get(url)

# Check if the response is successful
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # Scrape topics under 'A'
    topics = []
    for link in soup.find_all('a', class_="link-container table", href=True):
        topic_name_element = link.find('p', class_='heading text-underline')
        if topic_name_element:  # Check if the element exists
            topic_name = topic_name_element.get_text(strip=True)
            topic_link = link['href']
            topics.append({"name": topic_name, "link": topic_link})

    # Collect data for all topics
    data = []
    for topic in topics:
        print(f"Scraping content for: {topic['name']}")
        try:
            # Fetch the topic page HTML
            topic_response = requests.get(topic['link'])
            if topic_response.status_code == 200:
                topic_soup = BeautifulSoup(topic_response.text, 'html.parser')

                # Locate all content divs with the specific class
                content_divs = topic_soup.find_all('div', class_="sf_colsOut tabContent")

                # Initialize content dictionary for each topic
                content = {
                    "Overview": "N/A",
                    "Impacts": "N/A",
                    "WHO Response": "N/A"
                }

                # Map content divs to sections based on their availability
                if len(content_divs) > 0:
                    content["Overview"] = content_divs[0].get_text(strip=True, separator="\n")
                if len(content_divs) > 1:
                    content["Impacts"] = content_divs[1].get_text(strip=True, separator="\n")
                if len(content_divs) > 2:
                    content["WHO Response"] = content_divs[2].get_text(strip=True, separator="\n")

                # Append the data for this topic
                data.append({
                    "Health Topic": topic['name'],
                    "Link": topic['link'],
                    "Overview": content["Overview"],
                    "Impacts": content["Impacts"],
                    "WHO Response": content["WHO Response"]
                })
            else:
                print(f"Failed to fetch topic page for {topic['name']}. Status code: {topic_response.status_code}")
        except Exception as e:
            print(f"An error occurred while processing {topic['name']}: {e}")

    # Create a DataFrame
    df = pd.DataFrame(data)
    print(df)
else:
    print(f"Failed to fetch the main page. Status code: {response.status_code}")
```
