```python
import pandas as pd
```

```python
path ="C:/Users/Abdoulie Balisa/OneDrive\Desktop/schools.csv"
df = pd.read_csv(path)
df
```

```python
best_math_schools = df[df["average_math"] >= 640][['school_name', 'average_math']]
best_math_schools 
```

```python
best_math_schools = best_math_schools.sort_values(by="average_math", ascending=False)
best_math_schools
```

```python
# 2. Calculate total SAT and find the top 10 schools by combined SAT scores
df['total_SAT'] = df['average_math'] + df['average_reading'] + df['average_writing']
top_10_schools = df[["school_name", 'total_SAT']].sort_values(by='total_SAT', ascending=False).head(10)
top_10_schools 
```

```python
# 3. Calculate standard deviation of total SAT by borough
borough_stats = df.groupby('borough').agg(
    num_schools=('school_name', 'size'),
    average_SAT=('total_SAT', 'mean'),
    std_SAT=('total_SAT', 'std')
).reset_index()
```

```python
borough_stats.head(10)
```

```python
# Find the borough with the largest standard deviation
largest_std_dev = borough_stats.loc[borough_stats['std_SAT'].idxmax()]
largest_std_dev
```

```python
largest_std_dev
```

```python
largest_std_dev = pd.DataFrame({
    'borough': [largest_std_dev['borough']],
    'num_schools': [largest_std_dev['num_schools']],
    'average_SAT': [round(largest_std_dev['average_SAT'], 2)],
    'std_SAT': [round(largest_std_dev['std_SAT'], 2)]
})
```

```python
largest_std_dev
```

```python

# Print results
print("Best Math Schools:")
print(best_math_schools)

print("\nTop 10 Performing Schools by Total SAT:")
print(top_10_schools)

print("\nBorough with Largest Standard Deviation in SAT Scores:")
print(largest_std_dev)
```
